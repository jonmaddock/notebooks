```python
# Plot all solutions
data_dir = Path.cwd()
soln_files = list(data_dir.glob("*MFILE.DAT"))

# Order by number
soln_files = sorted(
    soln_files, key=lambda s: s.stem.removesuffix("_MFILE")
)

runs_metadata = []
for soln_file in soln_files:
    soln_name = soln_file.stem.removesuffix("_MFILE")
    runs_metadata.append(ps.RunMetadata(soln_file, soln_name))

solns =ps.plot_mfile_solutions(
    runs_metadata,
    "Global generic DEMO solutions relative to 1e-1_1e5 solution",
    percentage=True,
    relative_tag="1e-1_1e5",
    drop_f_values=False,
    plot_itvars=False,
)
```
